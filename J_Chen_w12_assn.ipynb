 ###  2. XML

```python
# Importing urllib.request
import urllib.request

# Importing objectify from the lxml library
from lxml import objectify
```

```python
# Reading the URL into path and headers
path, headers = urllib.request.urlretrieve('https://raw.githubusercontent.com/JinmingChen12/AIM-5001/master/W%2012/W12_XML')

parsed = objectify.parse(open(path))

# Getting the root
root = parsed.getroot()

# Creating an empty list data
data = []

# Using loop to create a dictionary of tag names
for elt in root.book:
    el_data = {}
    for child in elt.getchildren():
        el_data[child.tag] = child.pyval
    data.append(el_data)
    
# Showing the result xml 
xml = pd.DataFrame(data)

# Displaying books_xml
xml
```

### Transforming it to suitable for use in data ananlysis work

```python
# Using "json_normalize" to take the "books" and create a new Pandas dataframe that has the chosen information(attributes) which has "*"
trans = json_normalize(c['results'], 'books',  record_prefix ='*')

# Choosing the columns that might be suitable for use in data ananlysis work
DA = trans[['*rank', '*rank_last_week','*title', '*author', '*publisher','*description', '*weeks_on_list', '*contributor']]

DA
```

From the above data frame "DA", we can easier analyze some questions such as the relationship of book rank with each author, and how long the book been on the list. By analyzing this, we can give readers suggestion of choosing whose book to read in the future that might be more valuable. What is more, it is more organized than the "book_df" to do the analysis work.
